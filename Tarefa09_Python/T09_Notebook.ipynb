**I** - Faça um programa que pergunta o nome e a idade do usuario até que "quit" seja passado como nome do usuário. Salve todas as informações em um dicionário

```python
ages = dict()
name = ""
while True:
  name = input("Qual é o seu nome? (Digite 'quit' para sair): ")
  if "quit" == name.lower().strip(): 
    break
  age = input("Qual é a sua idade? ")
  ages.update({name: age})

print(ages)
```

In[1]: list('abcdefg')

**III** - Crie a classe `Calculadora` e a instancie. A calculadora deve possuir 4 métodos, um para cada operação básica.

```python
class Calculadora():
    """ Métodos para realizar as quatro operações básicas da matemática."""  

    
    def __init__(self):
        print("A calculadora está pronta para uso!")
        print("Utilize os métodos .adicao(), .subtracao(), .multiplicacao() e"+
              " .divisao() para realizar as operações básicas.")
  
  
    def adicao(self, *parcelas):
        total = sum(parcelas)
        return total
  
  
    def subtracao(self, minuendo, subtraendo):
        diferenca = minuendo - subtraendo
        return diferenca
  
  
    def multiplicacao(self, *fatores):
        produto = 1
        for f in fatores:
            produto *= f
        return produto
    
    
    def divisao(self, dividendo, divisor):
        quociente = dividendo / divisor
        return quociente
```

```python
# Instanciando a calculadora
my_calc = Calculadora()

# Teste de adição
print("3 + 7 + 5 = ", my_calc.adicao(3, 7, 5))

# Teste de subtração
print("4 - 3 = ", my_calc.subtracao(4, 3))

# Teste de multiplicação
print("7 * 6 = ", my_calc.multiplicacao(7, 6))

# Teste de divisão
print("20 / 5 = ", my_calc.divisao(20, 5))
```

**IV** - Crie a classe `Animal`, esta classe deve possuir os seguintes atributos:
* Número de patas
* Rabo
* Pêlos, penas ou escamas.
* Vertabrado ou invertebrado.
* Viviparos, oviparo ou ovoviviparos

Além disso, a classe deve possuir um método que mostra um texto com todas as informações do animal.

```python
class Animal():
    """Membro do reino Metazoa"""
    
    
    def __init__(self, patas: int, rabo: bool, metodo_reproducao: str, 
                 tipo_cobertura: str, vertebrado: bool):
        self.patas = patas
        self.rabo = rabo
        self.tipo_cobertura = tipo_cobertura.lower()
        self.vertebrado = vertebrado
        self.metodo_reproducao = metodo_reproducao.lower()
        
    def info(self):
        print("Este animal possui as seguintes características:")
        if self.patas > 0:
            print("    - Possui {} patas;".format(self.patas))
        else:
            print("    - Não possui patas;")
        if self.rabo:
            print("    - Possui rabo;")
        else:
            print("    - Não possui rabo;")
        print("    - É coberto por {};".format(self.tipo_cobertura))
        if self.vertebrado:
            print("    - É vertebrado;")
        else: 
            print("    - Não é vertebrado;")
        print("    - É {}.".format(self.metodo_reproducao))
        
        
```

```python
crocodilo = Animal(patas = 4, rabo = True, tipo_cobertura = "escamas", vertebrado = True, metodo_reproducao = "ovíparo")
```

```python
crocodilo.info()
```

**V** - Crie a classe `Mamífero` herdeira da classe `Animal`, esta classe deve possuir atributos que caracterizem bem os mamíferos.

Além disso, a classe deve possuir um método que mostra um texto com todas as informações do mamífero.  
PS.: Adicione pelo menos mais 2 atributos.

```python
class Mamifero(Animal):
    def __init__(self, patas: int, rabo: bool, metodo_reproducao: str,
                pares_mamas: int, habito_vida: str):
        assert metodo_reproducao.lower() in ["oviparo", "marsupial", "placentario"]
        super(Mamifero, self).__init__(patas, rabo, metodo_reproducao, 
                                       tipo_cobertura = "pelos", 
                                       vertebrado = True)
        self.pares_mamas = pares_mamas
        self.habito_vida = habito_vida.lower()
        
    def info(self):
        print("Este mamífero possui as seguintes características:")
        if self.patas > 0:
            print("    - Possui {} patas;".format(self.patas))
        else:
            print("    - Não possui patas;")
        if self.rabo:
            print("    - Possui rabo;")
        else:
            print("    - Não possui rabo;")
        print("    - É {};".format(self.metodo_reproducao))
        print("    - Possui {} mamas;".format(self.pares_mamas))
        print("    - É {}.".format(self.habito_vida))
```

**VI** - Crie uma classe `Loja` com os seguintes atributos:
* Nome
* Especialização
* Localização
* Horário de funcionamento
* Estoque
* Limite do Estoque
* Dinheiro

Sua loja deve vender produtos, comprar estoque e lucrar. Além disto, crie métodos que recebam um inteiro retornar se o estabelecimento esta funcionando.
Lembre-se que sua loja não pode estourar o limite de seu estoque.

```python
class Loja():
    """ Loja que compra, vende e apresenta o balanço """
    
    
    def __init__(self, nome: str, especializacao: str, localizacao: str, 
                 horario: tuple, limite_estoque: int, estoque: int = 0,
                 dinheiro: float = 100):
        self.nome = nome
        self.especializacao = especializacao
        self.localizacao = localizacao
        self.horario = horario
        self.limite_estoque = limite_estoque
        self.estoque = estoque
        self.dinheiro = dinheiro
        self.compras = 0
        self.vendas = 0
    
    
    def comprar(self, quantidade, preco = 10):
        if quantidade * preco > self.dinheiro:
            print("A loja não tem dinheiro para comprar essa quantidade de " +
                  "produtos!")
            return
        elif quantidade + self.estoque > self.limite_estoque:
            print("A loja não tem capacidade de receber todos esses podutos "+
                 "no estoque!")
            return
        self.dinheiro -= quantidade * preco
        self.estoque += quantidade
        self.compras += quantidade * preco
        print("Compra realizada com sucesso!")
        return
    
    
    def vender(self, quantidade, preco = 15):
        if quantidade > self.estoque:
            print("A loja não tem estoque suficiente para vender essa" + 
                  "quantidade de produtos!")
            return
        self.dinheiro += quantidade * preco
        self.estoque -= quantidade
        self.vendas += quantidade * preco
        print("Venda realizada com sucesso")
        return
        
        
    def balanco(self):
        print("Produtos em estoque: ", self.estoque)
        print("Dinheiro em caixa: R$", self.dinheiro)
        print("Total em compras: R$", self.compras)
        print("Total em vendas: R$", self.vendas)
        print("Resultado geral (lucro ou prejuízo): {} reais".format(
            self.vendas - self.compras))
        return
        
        
    def info_horario(self, horario: int):
        if horario >= self.horario[0] and horario < self.horario[1]:
            print("A loja {} se encontra em funcionamento!".format(self.nome))
        elif horario < self.horario[0]: 
            print("A loja {} ainda não abriu... Volte mais tarde!".format(
                self.nome))
        else: 
            print("A loja {} já fechou... Tente de novo amanhã!".format(
                self.nome))
```

```python
casas_paraiba = Loja(
    nome = "Casas Paraíba", especializacao = "Eletrodomésticos",
    localizacao = "R. das Pedras, 20", horario = (8, 20), limite_estoque = 12
)
```

```python
casas_paraiba.info_horario(7)
casas_paraiba.info_horario(9)
casas_paraiba.info_horario(21)
```

```python
casas_paraiba.comprar(20)
```

```python
casas_paraiba.comprar(10)
```

```python
casas_paraiba.vender(2)
```

```python
casas_paraiba.balanco()
```

**VII** - Crie uma função que recebe um inteiro `n` e retorna a sequência de Fibonnacci até o n-ésimo termo.

```python
def fibonacci(n):
    try:
        assert isinstance(n, int)
        assert n > 0
    except AssertionError:
        print("Erro! O número do termo deve ser um inteiro positivo!")
        return
    i = 2
    a = 0
    b = 1
    result = [1,]
    while n > 1 and i <= n:
        result.append(a + b)
        a = b
        b = result[-1]
        i += 1
    return(result)
```

**BÔNUS** - Crie uma função que recebe um inteiro `n` e retorna uma tupla com todos os números primos até n.

```python
def seq_primos(n):
    try:
        assert isinstance(n, int)
        assert n > 1
    except AssertionError:
        print("Erro! O número do termo deve ser um inteiro maior que 1!")
        return
    
    primos = list()
    i = 2
    while i <= n: # testa numero i, sendo i menor ou igual a n
        eh_primo = True # presume que i é primo
        for a in range(2, i): # para cada numero a, inteiro positivo e menor que i...
            if i % a == 0:     # ... testa se i é divisível por a
                eh_primo = False # se for, não é primo...
                break # já se pode parar o teste!
            else:
                pass # se não, deve-se continuar!
        if eh_primo: # se nenhuma divisão resultou em resto 0...
            primos.append(i) # ... então adicionar à lista de primos!
        i += 1 # e passar para o próximo teste na lista
        
    return(primos)
            
```
