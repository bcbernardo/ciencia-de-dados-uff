#### Faça uma nova coluna onde esta coluna seja uma função da coluna `age` de forma que a nova coluna deve conter os valores:
* Jovem: para as idades entre 0 e 24 anos
* Adulto: para o grupo de idades entre 25 e 74 anos
* Idoso: para 75+ anos de idade

Dica: Defina uma função e utilize o método `apply(foo)` para criar a nova coluna

```python
suicidios.age.unique()
```

```python
def faixa_idade(faixa):
    if faixa in ['5-14 years', '15-24 years']:
        return "Jovem"
    elif faixa in ['25-34 years', '35-54 years', '55-74 years']:
        return "Adulto"
    else:
        return "Idoso"
```

```python
suicidios.loc[:, "age_range"] = suicidios.age.apply(faixa_idade)
```

#### Com a nova coluna criada agrupe as variávies: País, ano, gênero e a nova coluna e calcule o a média dos suícidos.

```python
suicidios.groupby(["country", "year", "sex", "age_range"])["suicides_prop"].mean()
```
